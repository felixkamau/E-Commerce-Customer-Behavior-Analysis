## Segementation Analysis

```python
# Segement customers based on varibales
segment_data = df.groupby(['Gender','Age','City','Membership Type']).agg({
    'Total Spend': 'sum',
    'Items Purchased': 'sum',
    'Average Rating': 'mean',
}).reset_index()
```
